```python
# import packages
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# read data
suicide = pd.read_csv('suicide.csv')
```

```python
# split data
X = suicide.drop(columns=['country', 'sex', 'age', 'suicides_no','population', 'suicides/100k pop', 'country-year', 'generation'])

suicides_no = suicide['suicides_no']
X_train, X_test, suicides_no_train, suicides_no_test = train_test_split(X, suicides_no, test_size=0.2)
age = suicide['age']
X_train, X_test, age_train, age_test = train_test_split(X, age, test_size=0.2)
country = suicide['country']
X_train, X_test, country_train, country_test = train_test_split(X, country, test_size=0.2)
sex = suicide['sex']
X_train, X_test, sex_train, sex_test = train_test_split(X, sex, test_size=0.2)
```

```python
#Take input
year = input("Enter year: ")
```

```python
# plotting the points  
plt.plot(X, suicides_no) 
  
# naming the x axis 
plt.xlabel('Year') 
# naming the y axis 
plt.ylabel('Number of suicides') 
  
# giving a title to my graph 
plt.title('Chart showing the number of suicides from 1985 to 2015') 
  
# function to show the plot 
plt.show() 
```
